```python
import sys
import os
import torch
import cv2
import numpy as np
from PIL import Image
from torchvision import transforms
from matplotlib import pyplot as plt

# 1. Add the Tesseract-v2 directory to the Python Path
sys.path.insert(0, "/kaggle/working/Tesseract-v2")

# 2. Import custom modules (ensure you renamed preprocessing.py to xray_filters.py)
from inference_pipeline import load_pipeline_models, crop_with_padding
from xray_filters import preprocess_xray
from risk_scoring import compute_risk_score

# Configuration
YOLO_PATH = "/kaggle/working/models/yolo_best.pt"
CLS_PATH = "/kaggle/working/models/classifier_best.pth"
CLASS_NAMES = ["Baton", "Plier", "Hammer", "Powerbank", "Scissors", "Wrench", "Gun", "Bullet", "Sprayer", "HandCuffs", "Knife", "Lighter"]

SUPER_CLASS_MAP = {
    "Gun": "Lethal/Restraint", "Bullet": "Lethal/Restraint", "Knife": "Lethal/Restraint",
    "Baton": "Lethal/Restraint", "HandCuffs": "Lethal/Restraint", "Scissors": "Lethal/Restraint",
    "Hammer": "Metal Tool", "Wrench": "Metal Tool", "Plier": "Metal Tool",
    "Powerbank": "Benign Item", "Lighter": "Benign Item", "Sprayer": "Benign Item"
}

device = "cuda" if torch.cuda.is_available() else "cpu"

def run_test(image_path, output_filename="result.jpg"):
    # Load Models
    yolo, classifier, _, _ = load_pipeline_models(YOLO_PATH, CLS_PATH)
    
    # Read Image
    raw_img = cv2.imread(image_path)
    if raw_img is None:
        print(f"Error: {image_path} not found.")
        return

    # Create visualization copy
    vis_img = raw_img.copy()

    # Preprocess (The X-ray filter)
    processed_img = preprocess_xray(raw_img)
    proc_3ch = cv2.cvtColor(processed_img, cv2.COLOR_GRAY2BGR)

    # YOLO Inference
    results = yolo.predict(proc_3ch, conf=0.25, verbose=False)[0]
    
    print(f"\n--- Tesseract-v2 Inference Result ---")
    
    for box in results.boxes:
        # Coordinates
        xyxy = box.xyxy[0].cpu().numpy().astype(int)
        yolo_conf = float(box.conf[0].cpu().numpy())
        
        # Classifier Inference
        crop = crop_with_padding(raw_img, xyxy)
        crop_pil = Image.fromarray(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
        
        tf = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
        
        input_tensor = tf(crop_pil).unsqueeze(0).to(device)
        with torch.no_grad():
            logits = classifier(input_tensor)
            probs = torch.sigmoid(logits).squeeze()
            cls_idx = torch.argmax(probs).item()
            cls_prob = float(probs[cls_idx].item())

        det_class = CLASS_NAMES[cls_idx]
        super_cls = SUPER_CLASS_MAP.get(det_class, "Unknown")
        
        # Risk Score Calculation
        risk = compute_risk_score(yolo_conf, cls_prob, 0.2, class_name=det_class)
        
        # --- NEW: VISUALIZATION LOGIC ---
        # Color: BGR (Blue, Green, Red)
        if risk > 0.7:
            color = (0, 0, 255)    # RED
            status = "⚠️ THREAT"
        elif risk > 0.35:
            color = (0, 165, 255)  # ORANGE
            status = "NOTICE"
        else:
            color = (0, 255, 0)    # GREEN
            status = "CLEAR"

        # Draw Rectangle
        cv2.rectangle(vis_img, (xyxy[0], xyxy[1]), (xyxy[2], xyxy[3]), color, 3)
        
        # Draw Label Background
        label_text = f"{super_cls} ({risk:.2f})"
        (w, h), _ = cv2.getTextSize(label_text, cv2.FONT_HERSHEY_SIMPLEX, 0.7, 2)
        cv2.rectangle(vis_img, (xyxy[0], xyxy[1] - h - 10), (xyxy[0] + w, xyxy[1]), color, -1)
        
        # Put Text
        cv2.putText(vis_img, label_text, (xyxy[0], xyxy[1] - 5), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

        print(f"[{status}] {super_cls} ({det_class}) | Risk: {risk:.3f} | Conf: {cls_prob:.2%}")

    # Save to disk
    cv2.imwrite(output_filename, vis_img)
    print(f"\n✅ Output saved to: {output_filename}")

    # Display in Notebook
    plt.figure(figsize=(12, 8))
    plt.imshow(cv2.cvtColor(vis_img, cv2.COLOR_BGR2RGB))
    plt.title(f"Detection Results: {os.path.basename(image_path)}")
    plt.axis('off')
    plt.show()

if __name__ == "__main__":
    # Point this to a test image
    test_img = "/kaggle/input/x-ray-contraband-detection-dataset/data/images/test/PID_xray_04479.jpg"
    run_test(test_img)
```
